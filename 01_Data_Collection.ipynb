```python
ENGINE_MAP = {
    'Red Bull Racing' : 'RB-Ford',
    'RB' : 'RB-Ford',
    'Mercedes': 'Mercedes',
    'McLaren': 'Mercedes',
    'Williams': 'Mercedes',
    'Ferrari': 'Ferrari',
    'Haas': 'Ferrari',
    'Aston Martin': 'Honda',
    'Sauber': 'Audi',
    'Alpine': 'Mercedes',
    'Cadillac': 'Ferrari'

}
```

```python
def scrape_f1_history(start_year = 2023 , end_year=2025):
  master_data=[]

  for year in range(start_year , end_year +1):
    print(f"--- Processing Season : {year} ---")
    schedule = fastf1.get_event_schedule(year)

    races = schedule[schedule['EventFormat'] == 'conventional']

    for index , race in races.iterrows():
      race_name = race['EventName']
      print(f"Loading{race_name}")

      try:
        session = fastf1.get_session(year, race_name , 'R')
        session.load(laps = False , telemetry=False)
        results = session.results

        for _, driver in results.iterrows():
          master_data.append({
              'Year' : year,
              'Track' : race_name,
              'Driver' : driver['Abbreviation'],
              'Team' : driver['TeamName'],
              'Engine' : ENGINE_MAP.get(driver['TeamName'] , 'Other'),
              'Grid Position' : driver['GridPosition'],
              'FinishPosition' : driver['ClassifiedPosition'],
              'Status': driver['Status']})
      except Exception as e:
        print(f"Error loading {race_name} : {e}")

  return pd.DataFrame(master_data)

df = scrape_f1_history(start_year = 2023 , end_year=2026)
```

```python
df = scrape_f1_history()
# 'errors=coerce' turns things like 'R' or 'DNF' into NaN
df['FinishPosition'] = pd.to_numeric(df['FinishPosition'], errors='coerce')

Fill those NaNs (non-finishers) with 20 and convert to int
df['FinishPosition'] = df['FinishPosition'].fillna(20).astype(int);


save_path = '/content/drive/MyDrive/F1_2026_Project/data/f1_historical_cleaned.csv'
df.to_csv(save_path, index=False)

print(f"✅ Success! Dataset saved with {len(df)} rows.")


display(df.head())
```
